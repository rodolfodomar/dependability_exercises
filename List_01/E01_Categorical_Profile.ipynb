Exercise 1: Categorical Profile Analysis
This notebook analyzes the data from Table 1 of the exercise list, which contains the preferred PC brand among 50 respondents. We will perform the following steps:

1. Load the data into an R data frame.
2. Visualize the data using a bar chart and a pie chart.
3. Identify the mode of the dataset.
4. Provide a qualitative analysis of market dominance versus fragmentation.

**Setup and Data Preparation**
First, we import the necessary libraries and create a pandas DataFrame to hold the PC brand preferences from Table 1.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a dictionary with the data from Table 1
data = {
    'Brand': ['Dell', 'HP', 'Lenovo', 'Apple', 'Acer'],
    'Count': [18, 14, 9, 5, 4]
}

# Create a pandas DataFrame from the dictionary
pc_preferences_df = pd.DataFrame(data)

# Sort the DataFrame by 'Count' in descending order for better visualization
pc_preferences_df = pc_preferences_df.sort_values(by='Count', ascending=False)

# Display the DataFrame
print(pc_preferences_df)
```

01. (a) Visualizing the Data
Now we generate the required plots: a bar chart to compare the counts and a pie chart to visualize the market share. We will use seaborn for the bar chart and matplotlib for the pie chart.

```python
# --- Bar Chart (Corrected Code) ---
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Create the bar plot using hue for coloring and disabling the legend
bar_chart = sns.barplot(
    x='Brand',
    y='Count',
    data=pc_preferences_df,
    palette='viridis',
    hue='Brand', # <-- The fix is here
    legend=False  # <-- And here
)

# Add labels on top of each bar
# Note: The logic for adding text labels might need a small adjustment
# because using hue can change how bar positions are handled internally.
# A simple loop is more robust.
for i in range(pc_preferences_df.shape[0]):
    bar_chart.text(i, pc_preferences_df['Count'].iloc[i] + 0.2, 
                   pc_preferences_df['Count'].iloc[i], 
                   color='black', ha="center", va='bottom')


# Set titles and labels
plt.title('Preferred PC Brand Among 50 Respondents', fontsize=16)
plt.xlabel('Brand', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)

# Display the plot
plt.show()
```

```python
# --- Pie Chart ---
plt.figure(figsize=(8, 8)) # Set the figure size to make the circle look good

# Create the pie chart
plt.pie(
    pc_preferences_df['Count'],
    labels=pc_preferences_df['Brand'],
    autopct='%1.1f%%', # Automatically format the percentages
    startangle=140, # Rotate the chart for better aesthetics
    colors=sns.color_palette('viridis', len(pc_preferences_df)) # Use the same color palette
)

# Set title
plt.title('Market Share of Preferred PC Brands', fontsize=16)

# Ensure the pie chart is a circle
plt.axis('equal')

# Display the plot
plt.show()
```

01. (b) Identifying the Mode
The mode is the most frequently occurring category. In a frequency table like this, it's simply the category with the highest count. We can easily find this using pandas.

```python
# The mode is the brand in the first row of our sorted DataFrame
mode_brand = pc_preferences_df.iloc[0]['Brand']

print(f"The mode of the dataset is: {mode_brand}")
```

01. (c) Commentary on Dominance vs. Fragmentation
Based on the charts and data, we can analyze the market structure.

Dell is the clear leader with 36% of preferences (18 out of 50).

HP is the second major player with 28% (14 out of 50).

Together, these two brands account for 64% of the total preferences.

Conclusion: The distribution of preferences shows a clear pattern of dominance. The market is heavily concentrated around two main brands (Dell and HP), not fragmented among the various options. The remaining brands (Lenovo, Apple, Acer) occupy smaller, more niche segments within this sample.
